```python
from statsmodels.tsa.stattools import adfuller

def check_ADF_stationarity(df):
    """
    Performs the Augmented Dickey-Fuller test to check for stationarity.

    Parameters:
    - df: pandas DataFrame containing the time series data
    - column: str, the column to analyze

    Prints:
    - ADF Statistic
    - p-value
    - Critical values
    - Stationarity conclusion
    """
    result = adfuller(df)

    print(f"ADF Statistic: {result[0]:.4f}")
    print(f"p-value: {result[1]:.4f}")
    print("Critical Values:")
    for key, value in result[4].items():
        print(f"   {key}: {value:.4f}")

    if result[1] < 0.05:
        print("✅ The time series is **stationary** (rejects H0).")
        return 1
    else:
        print("❌ The time series is **non-stationary** (fails to reject H0).")
        return 0


```

### train test split

```python

import datetime
from dateutil.relativedelta import relativedelta  # Import relativedelta

# Define the train end date
train_end_date = datetime.datetime(1992, 12, 1)

# Calculate the test start date by adding 1 month
test_start = train_end_date + relativedelta(months=1)

train_data=df[:train_end_date]
test_data=df[test_start:]
```

```python
# Convert selected districts into DataFram
col="Bankura_temp"
selected_districts=dtw_result_df[col].dropna().index
exog_train = pd.DataFrame({k: train_data[k] for k in selected_districts}) if len(selected_districts)!=0 else None 
exog_test = pd.DataFrame({k: test_data[k] for k in selected_districts})if len(selected_districts)!=0 else None 
```

### sarima model

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima import auto_arima
from sklearn.metrics import r2_score
predict_result=test_data.copy()

pred_start_date = test_data.index[0]
pred_end_date = test_data.index[-1]

my_order = (2,0,6)
my_seasonal = (2, 0, 1, 12)
for i,colum in enumerate(train_data.columns):
    print(colum)
    # auto_model = auto_arima(df[colum], seasonal=True, m=12, stepwise=True, trace=True)
    selected_districts=dtw_result_df[colum].dropna().index
    exog_train = pd.DataFrame({k: train_data[k] for k in selected_districts}) if len(selected_districts)!=0 else None 
    print("get_model")
    model = SARIMAX(train_data[colum],exog=exog_train, order=my_order, seasonal_order=my_seasonal)
    # model = SARIMAX(train_data[colum], order=auto_model.order, seasonal_order=auto_model.seasonal_order)
    
# Fit the model
    model_fit = model.fit()
    exog_test = pd.DataFrame({k: test_data[k] for k in selected_districts})if len(selected_districts)!=0 else None 
    
    pred = model_fit.predict(start=pred_start_date,exog=exog_test, end=pred_end_date)
    predict_result[colum]=pred
    print(pred)
    print(f" r2 score of {colum}: {r2_score(test_data[colum],pred)}")
    
    
    
    
    
```
